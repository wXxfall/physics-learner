````python
import numpy as np
import matplotlib.pyplot as plt

# 定义参数
Lambda = 5  # 势阱深度参数

# 定义函数
def f1(y):
    return y * np.tan(y)

def g1(y):
    # 计算余切：cot(y) = 1/tan(y)，避免除零
    with np.errstate(divide='ignore', invalid='ignore'):
        cot_y = 1 / np.tan(y)
        cot_y[np.isinf(cot_y)] = 0  # 处理奇点
    return -y * cot_y

def f2(y):
    return np.sqrt(Lambda**2 - y**2)

def f3(y):
    return np.sqrt((2 * Lambda)**2 - y**2)

def f4(y):
    return np.sqrt((3 * Lambda)**2 - y**2)

def f5(y):
    return np.sqrt((4 * Lambda)**2 - y**2)

def f6(y):
    return np.sqrt((5 * Lambda)**2 - y**2)

# 生成数据点
y = np.linspace(0.01, 5 * Lambda, 1000)  # 从0.01开始避免奇点

# 计算函数值
f1_values = f1(y)
g1_values = g1(y)
f2_values = f2(y)
f3_values = f3(y)
f4_values = f4(y)
f5_values = f5(y)
f6_values = f6(y)

# 处理复数结果（当根号内为负时）
f2_values = np.real(f2_values)
f3_values = np.real(f3_values)
f4_values = np.real(f4_values)
f5_values = np.real(f5_values)
f6_values = np.real(f6_values)

# 绘制图像
plt.figure(figsize=(10, 8))

# 绘制函数曲线
plt.plot(y, f1_values, 'b-', linewidth=2, label='偶宇称')
plt.plot(y, g1_values, 'g-', linewidth=2, label='奇宇称')
plt.plot(y, f2_values, 'r-', linewidth=2)
plt.plot(y, f3_values, 'r-', linewidth=2)
plt.plot(y, f4_values, 'r-', linewidth=2)
plt.plot(y, f5_values, 'r-', linewidth=2)
plt.plot(y, f6_values, 'r-', linewidth=2)

# 设置图像属性
plt.xlim(0, 5 * Lambda)
plt.ylim(0, 5 * Lambda)
plt.xlabel('y')
plt.ylabel('f(y)')
plt.title('有限深方势阱能量量子化')
plt.legend()
plt.grid(True)

# 显示图像
plt.show()
````
