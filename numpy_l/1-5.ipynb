np.arange() 生成一个等维数组,默认步长为1,可以指定步长.用法示例：
```
np.arange(start,stop,step)
```

````python
data3 = np.arange(10,49)
print(data3)
````

反转第一个向量：data3[::-1]用法示例：
```
data3[start:stop:step]
```

````python
#反转第一个向量
data3 = data3[::-1]#从后往前反转
print(data3)
````
