````python
#use tile function toact create a 8*8 matrix with 1 and 0

a = np.tile(np.array([[0,1],[1,0]]),(4,4))
matrix.print_matrix(a)
````

 np.tile 函数用于重复数组

 第一个参数 np.array([[0,1],[1,0]]) 是基础数组，一个 2x2 的矩阵

第二个参数 (4,4) 表示在第一个维度（行）重复 4 次，在第二个维度（列）重复 4 次

 最终生成一个 8x8 的矩阵（2*4=8 行，2*4=8 列）

# 特征归一化（Min-Max 归一化）
- 方法 ：将特征值缩放到 [0,1] 范围
- 公式 ： $x' = \frac{x - x_{min}}{x_{max} - x_{min}}$
- 应用 ：神经网络输入、需要范围约束的算法

$\sqrt{\sum_{i=1}^{n} x_i^2}$


````python
#归一化：
b = np.random.random((5,5))
data_max, data_min = np.max(b), np.min(b)
b = (b - data_min) / (data_max - data_min)
matrix.print_matrix(b)
#验证归一化，求矩阵的模长
norm = np.linalg.norm(b)#矩阵的模长,原理是将矩阵的每个元素平方后平方和再开平方根：
print(norm)
````

计算轴（axis） 是 NumPy 中用于指定 沿着哪个维度进行计算 的参数，它是理解 NumPy 数组操作的核心概念。

### 轴的定义
- 轴是数组的维度 ：
  - 一维数组只有 1 个轴（axis=0）
  - 二维数组有 2 个轴（axis=0：行方向，axis=1：列方向）
  - 三维数组有 3 个轴（axis=0：深度，axis=1：行，axis=2：列）
- 负数表示从末尾计数 ： axis=-1 表示 最后一个轴

````python
#如何从零位对浮点数做舍入
data  = np.random.uniform(-10,+10,10)
print(np.copysign(np.ceil(np.abs(data)),data))
#np.abs(data) 作用 ：计算数组 data 中每个元素的 绝对值
#np.ceil(...)  作用 ：对数组中的每个元素进行 向上取整 （向正无穷方向取整）示例 ： [1.2, 3.4, 5.6] → [2.0, 4.0, 6.0]
````

3. np.copysign(a, b)
- 作用 ：将 b 的符号复制给 a ，即结果的数值部分是 a ，符号与 b 相同
- 参数 ：
  - a ：数值部分（这里是向上取整后的结果）
  - b ：符号来源（这里是原始 data 数组）
- 示例 ：
  - a=[2.0, 4.0, 6.0] ， b=[-1.2, 3.4, -5.6]
  - 结果： [-2.0, 4.0, -6.0]

```np.random.uniform``` 是 NumPy 中用于 生成均匀分布随机数 的函数
```
np.random.uniform(low=0.0, high=1.0, size=None)
```
### 数解释
- low ：均匀分布的 最小值 （包含），默认为 0.0
- high ：均匀分布的 最大值 （不包含），默认为 1.0
- size ：生成随机数的 数量或形状 ，可以是整数（生成一维数组）或元组（生成多维数组）
### 核心特点
- 均匀分布 ：在 [low, high) 范围内的每个值出现的概率相等
- 浮点数 ：生成的是浮点数随机数
- 连续分布 ：不同于 randint 生成的整数， uniform 生成的是连续区间内的随机数
