### 语法解释
- e ：NumPy 数组变量
- [1:-1, 1:-1] ：二维切片操作

    第一个维度（行）： 1:-1 表示从第2行到倒数第2行（不包括最后一行）
    
    第二个维度（列）： 1:-1 表示从第2列到倒数第2列（不包括最后一列）

```python
# 创建 10×10 的全1数组
e = np.ones((10,10))  # np.ones(shape) 生成指定形状的全1数组
# 将数组内部区域设置为0，保持边界为1
e[1:-1,1:-1] = 0  # 切片操作：1:-1 表示从第2个元素到倒数第2个元素，不包括边界
print('e=')  # 打印数组标签
matrix.print_matrix(e)  # 以矩阵风格打印处理后的数组
```

### 参数解释：
- f ：输入数组，需要被填充的原始数组
- pad_width=1 ：填充宽度，指定在数组每个维度的边缘添加的元素数量。值为 1 表示在每个维度的前后各添加 1 层元素
- mode='constant' ：填充模式，'constant' 表示使用常量值填充
- constant_values=0 ：当 mode 为 'constant' 时使用的填充值，这里设置为 0
### 其他填充模式：
- 'edge' ：使用数组边缘值填充
- 'reflect' ：镜像反射填充
- 'wrap' ：循环填充
- 'symmetric' ：对称填充

```python
#对于一个数组，如何用0填充边界
f = np.ones((3,8))
f = np.pad(f,pad_width = 1,mode = 'constant',constant_values = 0)
matrix.print_matrix(f)

```

np.nan 是 NumPy 中表示"不是一个数字"（Not a Number）的特殊值，它具有以下特点：

### 定义与本质
- 全称 ：Not a Number（非数字）
- 类型 ：特殊的浮点数
- 标准 ：遵循 IEEE 754 浮点数标准
### 核心特性
1. 不等于自身 ： np.nan == np.nan 返回 False ，这是它最独特的特性
2. 传播性 ：任何与 np.nan 的运算结果都是 np.nan
3. 比较结果 ：任何与 np.nan 的比较运算（>, <, ==, !=）都返回 False
4. 表示缺失值 ：常用于表示数据中的缺失或无效值

```python
# 任何数与 np.nan 相乘结果都是 np.nan（Not a Number）
print("0 * np.nan:", 0 * np.nan)  # 结果: nan
# np.nan 是唯一一个不等于自身的值，这是 IEEE 754 浮点数标准的规定
print("np.nan == np.nan:", np.nan == np.nan)  # 结果: False
# 任何与 np.nan 的比较运算结果都是 False
print("np.inf > np.nan:", np.inf > np.nan)  # 结果: False
# 任何涉及 np.nan 的运算结果都是 np.nan
print("np.nan - np.nan:", np.nan - np.nan)  # 结果: nan
# 浮点数精度问题：3 * 0.1 实际结果不是精确的 0.3
print("0.3 == 3 * 0.1:", 0.3 == 3 * 0.1)  # 结果: False
# 显示 3 * 0.1 的实际计算结果
print("3 * 0.1:", 3 * 0.1)  # 结果: 0.30000000000000004
```

1. 第二个参数 ： k = -1
   
   - k 表示对角线的位置
   - k = 0 ：主对角线（默认值）
   - k > 0 ：主对角线以上的对角线（如 k=1 是主对角线上方第一条）
   - k < 0 ：主对角线以下的对角线（如 k=-1 是主对角线下方第一条）
### 函数用法
np.diag(v, k=0) 函数的作用：

- 当 v 是 一维数组 时：创建一个方阵，将 v 的元素放在指定对角线 k 上，其余位置为0
- 当 v 是 二维数组 时：提取 v 中指定对角线 k 的元素，返回一维数组

```python
g = np.diag(1 +np.arange(4),k = -1)
matrix.print_matrix(g)
```
