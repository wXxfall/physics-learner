### 语法解释
- e ：NumPy 数组变量
- [1:-1, 1:-1] ：二维切片操作

    第一个维度（行）： 1:-1 表示从第2行到倒数第2行（不包括最后一行）
    
    第二个维度（列）： 1:-1 表示从第2列到倒数第2列（不包括最后一列）

```python
e = np.ones((10,10))
e[1:-1,1:-1] = 0
print('e=')
matrix.print_matrix(e)
```
