````python
#如何得到昨天明天的的日期
# np.datetime64('today') 获取当前日期
#  np.timedelta64(1, 'D') 表示 1 天的时间间隔;两者相加得到明天的日期
yesterday = np.datetime64('today') - np.timedelta64(1, 'D')

print(yesterday)
today = np.datetime64('today')
print(today)
tomorrow = np.datetime64('today') + np.timedelta64(1, 'D')
print(tomorrow)
````

``` np.timedelta64(1, 'D') ```  是 NumPy 中用于表示 时间间隔 的函数，具体含义如下：
- np.timedelta64 ：NumPy 中表示时间间隔的类型
- 第一个参数 1 ：时间间隔的数量
- 第二个参数 'D' ：时间单位，表示 天（Day）

````python
date = np.arange("2016-07","2016-08",dtype="datetime64[D]")#start,stop,step
date
````
