```np.sum``` 和 ```np.add.reduce``` 是等价的
```np.sum``` 是一个更高级的函数，它可以对数组进行求和操作，而 ```np.add.reduce``` 是一个更基础的函数，它对数组进行加法操作，最后返回一个标量结果。

````python
data = np.arange(10,49)
np.add.reduce(data)
print(np.add.reduce(data))
````

````python
#将直角坐标系的一个矩阵变化到极坐标
data = np.random.random((10,2))
x,y=data[:,0],data[:,1]
R = np.sqrt(x**2+y**2)
print(R)
theta = np.arctan2(y,x)
print(theta)
print(theta*180/np.pi)
````
